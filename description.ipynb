```python
import pandas as pd
```

```python
URL = "https://resources.lendingclub.com/LCDataDictionary.xlsx"
```

```python
def description(URL=URL):
    return pd.read_excel(
        URL, 
        sheet_name="browseNotes", 
        engine="openpyxl", 
        usecols=[0, 1], 
        skipfooter=2
    )

desc = description()
```
